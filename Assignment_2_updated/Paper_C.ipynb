## Prediction of appliances energy use in smart homes

### The flow of this paper
Purpose: How to predict energy consumption of different appliances in homes ---> Why we need energy prediction in housing? ---> Introduction of four different predictors ---> How to assess the performance of predictors and improve the prediction precision   
<font color='red'>In this paper, the predictor aims at predicting whether the service will consume energy or not during each hour of the next 24h.</font>

### 1. Some important concept

Two processings are proposed to improve the performance of the predictor: segmentation and aggregation.   
A household energy control system with three layers: 
+ anticipative layer( is responsible for the prediction of energy comsumption)
+ reactive layer
* device layer   


### 2. Smart grid

* Demand response: ability to cut off some electrical loads at peak times to ameliorate the need for peaking generation sources.  
* Demand dispatch: ability to control in a precise manner the individual loads at all times, not only at peak period.   
There are two types of load management:
+ Direct control: increase the energy production in case of higher load demand 
+ Control by cost: increase the energy price at peak period to decrease the energy consumption at that times(but the total consumption is the same)

### 4. Four different predictors

#### 4.1 The "will always consume" predictor
It considers that the appliance will consume energy permanently.

#### 4.2 The "will never consume" predictor
It assumes that the service will not consume at all in the next day.   
<font color='red'>The value of precision for this predictor is the complement of the "will always consume" predictor precision.</font>

#### 4.3 The ARMA predictor
It was used in order to predict the next day energy consumption.

#### 4.4 The proposed predictor
This predictor model the energy consumption as a stochastic process.
