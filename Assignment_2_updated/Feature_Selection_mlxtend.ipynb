```python
import pandas as pd
import numpy as np
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor
from sklearn.neural_network import MLPRegressor
from sklearn import linear_model
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error
from itertools import chain, combinations
from sklearn.model_selection import cross_val_score
from mlxtend.feature_selection import SequentialFeatureSelector as SFS
```

### 1.Read data

```python
def read_and_get_data():
    train_data = pd.read_csv('./training.csv')
    test_data = pd.read_csv('./testing.csv')

    x_train = train_data.values[:,2:28]
    y_train = train_data['Appliances'].values

    x_test = test_data.values[:,2:28]
    y_test = test_data['Appliances'].values
    return x_train, y_train, x_test, y_test

def read_and_get_data1():
    train_data = pd.read_csv('./training.csv')
    test_data = pd.read_csv('./testing.csv')

    x_train = train_data.drop(['Appliances'], axis = 1).values
    y_train = train_data['Appliances'].values

    x_test = test_data.drop(['Appliances'], axis = 1).values
    y_test = test_data['Appliances'].values
    return x_train, y_train, x_test, y_test
```

### 2. forward search and results

```python
rf = RandomForestRegressor(max_depth=100)
x_train, y_train, x_test, y_test = read_and_get_data()

sfsSelector = SFS(rf,
                  k_features=15,
                  forward=True,
                  scoring='r2',
                  cv=5)
```

```python
sfsSelector = sfsSelector.fit(x_train, y_train)
```
